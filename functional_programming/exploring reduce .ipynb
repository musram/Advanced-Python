```python
import numpy as np
from functools import reduce
```

REDUCE

```python
a = np.random.randint(100, size=100)
                      
```

```python
a = a.tolist()
```

list


```python
reduce(lambda acc, val : acc + [val], a, [])
```

set

```python
reduce(lambda acc, val : acc.union({val}), a, set())
```

dict

```python
reduce(lambda acc, val : {**acc, **{val : acc.get(val, 0) + 1}}, a, {})
       
```

```python
def count_freq(acc, val):
    acc[val] = acc.get(val, 0) + 1
    return acc
reduce(count_freq, a, {})
```
